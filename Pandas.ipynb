```python
import pandas as pd
import numpy as np
```

```python
data=[
      [1,2,3,'M'],
      [np.nan,5,6,np.nan],
      [7,np.nan,8,'F'],
      [np.nan,np.nan,np.nan,'F']
]
```

```python
data
```

```python
dframe=pd.DataFrame(data,columns=['a','b','c','d'])
dframe
```

```python
dframe['a']
```

```python
# fillna
dframe.fillna(0)
```

```python
# find avg of column a.It is nelecting null values from that column
avg_a=dframe['a'].mean() # for numrical value
avg_a
```

```python
avg_b=dframe['b'].mean()
avg_b
```

```python
avg_c=dframe['c'].mean()
avg_c
# give to many decimal point so use round ()
```

```python
avg_c=round(dframe['c'].mean(),1)
avg_c
# give to many decimal point so use round ()
```

```python
dframe.shape[1] # it wiill gives the rows value 
```

```python
dframe.shape
```

```python
avg_d=dframe['d'].mode()
avg_d
```

```python
d_mode=dframe['d'].mode()[0] #for catgorical Value
d_mode
```

```python
dframe.fillna({'a':avg_a,'b':avg_b,'c':avg_c,'d':d_mode},inplace=True)
```

```python
dframe
```

 **Connecting Dataframes**

```python
df1 = pd.DataFrame([[1,2,3],[4,5,6]],columns=['a','b','c'])
df1
```

```python
df2 = pd.DataFrame([[10,20,30],[40,50,60]],columns=['a','b','c'])
df2
```

```python
df3 = pd.DataFrame([[11,22],[33,44]],columns=['c1','c2'])
df3
```

```python
d1 = {'movieid':[1,2,3,4,5],
      'title':['Leo','Jawan','Gadar','Pathan','Uri']}
movies = pd.DataFrame(d1)
```

```python
import numpy as np
d2 = {'memid':[1,2,3,4],
      'name':['Tom','Tim','Jon','Jim'],
      'movieid':[1,np.nan,3,5]}
members = pd.DataFrame(d2)
```

```python
#concat
# for records
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2],ignore_index=True) #ignore_index=True will reset the index in sequential manner
```

```python
df3
```

```python
# join() : for columns
```

```python
df1.join(df3)
```

```python
# merge() :it perform operation like sql joins
```

```python
movies
```

```python
# inner join
pd.merge(movies,members, on='movieid',how='inner') # if how= not specified then also inner join will be done by default
```

```python
pd.merge(movies,members, on='movieid',how='left')
```

```python
pd.merge(movies,members, on='movieid',how='right')
```

```python
pd.merge(movies,members, on='movieid',how='outer')
```
