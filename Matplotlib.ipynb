```python
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# Create a figure
plt.figure(figsize=(10, 6))
```

```python
# 1. Line plot
plt.plot(x, y1, 'r--', label="sin(x)")   # red dashed line
plt.plot(x, y2, 'b-', label="cos(x)")    # blue solid line
```

```python
# 2. Scatter plot
plt.scatter(x, y1, color="green", s=20, alpha=0.6, label="scatter sin(x)")
```

```python
# 3. Labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Matplotlib Example: Line, Scatter, Labels, Legend")
```

```python
# 4. Grid
plt.grid(True, linestyle="--", alpha=0.7)
```

```python
# 5. Legend
plt.legend()
```

```python
# Show plot
plt.show()
```

```python
# 6. Bar chart
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]
```

```python
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color="orange")
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# 7. Histogram
data = np.random.randn(1000)  # random normal distribution
plt.figure(figsize=(6, 4))
plt.hist(data, bins=30, color="purple", alpha=0.7)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# 8. Subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, y1, 'r-')
axs[0, 0].set_title("sin(x)")

axs[0, 1].plot(x, y2, 'b-')
axs[0, 1].set_title("cos(x)")

axs[1, 0].scatter(x, y1, color="green")
axs[1, 0].set_title("Scatter sin(x)")

axs[1, 1].hist(data, bins=20, color="purple")
axs[1, 1].set_title("Histogram")

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 10, 500)
y = np.sin(x)
z = np.exp(-0.1 * x) * np.sin(x)   # damped sine wave
```

```python
# 1. Multiple plots with twin y-axis
fig, ax1 = plt.subplots(figsize=(8,5))
ax2 = ax1.twinx()   # second y-axis

ax1.plot(x, y, 'g-', label="sin(x)")
ax2.plot(x, z, 'b--', label="damped sin(x)")

ax1.set_xlabel("X-axis")
ax1.set_ylabel("sin(x)", color="g")
ax2.set_ylabel("damped sin(x)", color="b")

plt.title("Twin Y-axis Example")
plt.show()
```

```python
# 2. Fill between curves
plt.figure(figsize=(8,5))
plt.plot(x, y, 'r-', label="sin(x)")
plt.fill_between(x, y, 0, where=(y > 0), color="c", alpha=0.3, label="positive area")
plt.fill_between(x, y, 0, where=(y < 0), color="orange", alpha=0.3, label="negative area")
plt.legend()
plt.title("Fill Between Example")
plt.show()
```

```python
# 3. Polar plot
theta = np.linspace(0, 2*np.pi, 500)
r = 1 + np.sin(6*theta)
plt.figure(figsize=(6,6))
ax = plt.subplot(111, projection='polar')
ax.plot(theta, r, color="purple")
ax.set_title("Polar Plot Example")
plt.show()
```

```python
# 4. 3D Surface plot
from mpl_toolkits.mplot3d import Axes3D

X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap="magma")
ax.set_title("3D Surface Plot Example")
plt.show()
```
