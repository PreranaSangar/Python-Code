```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load sample dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.histplot(tips["total_bill"], bins=20, kde=True)
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True)
plt.show()
```

```python
sns.pairplot(tips, hue="sex")
plt.show()
```

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="magma")
plt.show()
```

```python
sns.regplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.kdeplot(data=tips, x="total_bill", fill=True)
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=tips, estimator="mean", ci=None)
plt.show()
```

```python
sns.countplot(x="day", data=tips)
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="sex", style="smoker", size="size", data=tips)
plt.show()
```

```python
sns.lineplot(x="size", y="tip", data=tips, estimator="mean")
sns.set_palette("magma")
plt.show()
```

```python
g = sns.FacetGrid(tips, col="sex", row="time")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue="sex", col="day", data=tips)
plt.show()
```

```python
sns.set_style("whitegrid")   # options: dark, white, ticks, darkgrid
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.set_palette("Dark2")
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
ax = sns.scatterplot(x="total_bill", y="tip", data=tips)
sns.set_palette("pink")
ax.set_title("Bill vs Tip")
ax.set_xlabel("Total Bill Amount")
ax.set_ylabel("Tip Amount")
plt.show()
```
